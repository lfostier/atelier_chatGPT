```python
# Normaliser les noms de colonnes pour uniformiser 'country' ou 'Country' en 'country'
df_mosquitoes.columns = [col.lower() for col in df_mosquitoes.columns]
df_temp.columns = [col.lower() for col in df_temp.columns]
df_pluie.columns = [col.lower() for col in df_pluie.columns]

# Extraire les ensembles de pays de chaque DataFrame
countries_mosquitoes = set(df_mosquitoes['country'].unique())
countries_temp = set(df_temp['country'].unique())
countries_pluie = set(df_pluie['country'].unique())

# Vérifier si les ensembles de pays sont identiques
if countries_mosquitoes == countries_temp == countries_pluie:
    print("Les trois DataFrames ont les mêmes pays.")
else:
    print("Les DataFrames n'ont pas les mêmes pays.")
    # Afficher les différences
    print("Pays dans df_mosquitoes mais pas dans df_temp :", countries_mosquitoes - countries_temp)
    print("Pays dans df_mosquitoes mais pas dans df_pluie :", countries_mosquitoes - countries_pluie)
    print("Pays dans df_temp mais pas dans df_mosquitoes :", countries_temp - countries_mosquitoes)
    print("Pays dans df_temp mais pas dans df_pluie :", countries_temp - countries_pluie)
    print("Pays dans df_pluie mais pas dans df_mosquitoes :", countries_pluie - countries_mosquitoes)
    print("Pays dans df_pluie mais pas dans df_temp :", countries_pluie - countries_temp)
```
